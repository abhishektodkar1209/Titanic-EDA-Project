# **Survived vs Age**

```python
# kde : Kernel Density Estimation
```

```python
df[df['Survived'] == 0]['Age']
```

```python
 sns.kdeplot(df[df['Survived'] == 0]['Age'],color='black', label='Not Survived')
sns.kdeplot(df[df['Survived'] == 1]['Age'],color='green', label='Survived')
plt.legend()
plt.show()
```

# **survived** vs **fare**

```python
sns.kdeplot(df[df['Survived'] == 0]['Fare'],color='black', label='Not Survived')
sns.kdeplot(df[df['Survived'] == 1]['Fare'],color='green', label='Survived')
plt.legend()
plt.show()
```
